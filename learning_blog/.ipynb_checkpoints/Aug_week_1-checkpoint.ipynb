# 相关知识点
### 蛋白质结构

**蛋白质结构**是指蛋白质分子的空间结构。
1. 蛋白质主要有碳，氢，氧，氮，硫等化学元素组成。
2. 蛋白质都是由20种不同的**L型$\alpha$氨基酸**链接形成的**多聚体**
3. 在形成蛋白质后，这些氨基酸被称为**残基**。

**蛋白质和多肽**
1. 蛋白质和多肽之间的界限并不是很清晰，有人基于发挥功能性作用的结构域所需的残基数认为，若残基数少于40，被称为多肽或肽。

**氨基酸**
1. 是生物学上重要的有机化合物，由**氨基**(-$NH_2$)和**羧基**(-COOH)的**官能团**，以及连到每一个氨基酸的侧链组成
2. 氨基酸是构成蛋白质的基本单位，赋予蛋白质特定的分子结构形态，使分子具有生化活性（biochemical activity）
3. 蛋白质是生物体重要的生物活性分子（bioactive molecule），包括催化新城代谢的酶（又称酵素）

**聚合物**
1. 又称多聚体，是指具有非常大的分子量的化合物

**转录**

1. 转录是遗传信息从DNA流向RNA的过程。即以双链DNA中的确定的一条链（模版链用于转录，编码链不用于转录）模版，以A,U,C,G四种核糖核苷酸为原料，在RNA聚合酶催化下合成RNA的过程。
2. 作为蛋白质生物合成的第一步，进行转录时，一个基因会被读取并被复制为mRNA，即特定的DNA片段作为遗传信息模版，以依赖DNA的RNA聚合酶作为催化剂，通过碱基互补的原则合成前体mRNA。RNA聚合酶通过与一系列组成构成动态复合体，完成转录起始，延伸，终止等过程。
3. 生成的mRNA写有的密码子，进入核糖体后可以实现蛋白质的合成。转录仅以DNA的一条链作为模版，被选为模版的单链称为模版链，亦称无义链；另一条单链称为非模版链，即编码链，因编码链与转录生成的RNA序列T变为U外其他序列一致，所以又称有义链。DNA上的转录区域称为转录单位。

**转录起始**

RNA聚合酶正确识别DNA模版链上的启动子并形成由酶，DNA和核苷三磷酸(NTP)构成的三元起始复合物，转录即自此开始。第一个核苷三磷酸与第二个核苷三磷酸缩合生成3'-5'磷酸二酯键后，则启动阶段结束，进入延伸阶段。

**RNA（核糖核酸） 和DNA（脱氧核糖核酸）**
1. DNA是双链，RNA是单链
2. RNA含有四种碱基，A(腺嘌呤）,U（尿嘧啶）,G（鸟嘌呤）,C（胞嘧啶）
3. DNA将尿嘧啶替换成T(胸腺嘧啶)
4. RNA分为多种类型
    1. mRNA（传讯RNA)为遗传讯息的传递者，能够知道蛋白质的合成，因为它有编码蛋白质的能力，也被称为编码RNA。
    2. 而其他没有编码蛋白质的能力的RNA被称为非编码RNA(ncRNA). 这些经由催化生化反应，或透过调控或参与基因表达过程发挥相应的生理功能,比如
        1. tRNA（转运RNA)在翻译过程中起转运RNA的作用
        2. rRNA（核糖体RNA)于翻译过程中起催化肽链形成的作用
        3. sRNA（小RNA）起到调控基因表达的作用。
5. RNA通常由DNA经由转录生成。

**promoter（启动子）**
启动子是RNA聚合酶识别，结合和开始转路的一段DNA序列，

**RNA polymerase（RNA聚合酶）**
1. RNA聚合酶是以一条DNA链或RNA为模版，三磷酸核糖核苷为底物，通过磷酸二酯键而聚合的合成RNA的酶，因为在细胞内与基因DNA的遗传信息转录为RNA有关，所以也称转录酶。
2. 催化RNA的合成，负责从DNA模版制造RNA的酶

**基因表达**

基因表达是指细胞在生命过程中，把储存在DNA顺序中遗传信息经过转录和翻译，转变成具有生物活性的蛋白质分子


**bacteria（细菌）的RNA聚合酶**
1. 细菌是生物的主要类群之一，属于细菌界，是所有生物中数量最多的一类。
2. 在细菌中，相同的酶催化三种RNA的合成：mRNA, rRNA, tRNA.
3. RNA聚合酶是相对大的分子。核心酶有5个亚基
    1. 两个$\alpha$亚基组合成酶及辨认调节因子。每个亚基有两个区，$\alpha$C末端区及$\alpha$N末端区，分别与启动子结合及与聚合酶的其他部分结合
    2. $\beta$: 有着聚合酶的活动，负责催化RNA的合成
    3. $\beta'$: 与DNA结合
    4. w:
    
4. 为与启动子的特定区域结合，核心酶须有其他亚基，称为$\sigma$，$\sigma$因子大大减低RNA聚合酶与非特定的DNA的关系。

### abstract
#### background
蛋白质生物学的核心是结构怎么产生对应的作用。但是很多方法的效率取决于蛋白质结构的表达方式。这些一般都是多用途的，不会因为某个应用特意优化。

他们用了3DCNN方法，能够自动的从未处理的原子分布中提取出针对特定任务的特征。作为试点试验，他们用这个模型分析了20个氨基酸的周围的蛋白质微环境，然后预测跟蛋白质结构最兼容的氨基酸。为了进一步验证我们的模型，我们从预测统计数据中创建了两个氨基酸替换矩阵并且用他们来预测T4溶酶菌结构的突变效果。

#### results
我们的深度3DCNN在预测精确度上比传统的手动设计特征方法高了两倍，并且成功的概括了相似和不同的微环境的已知信息。根据我们的预测和替换矩阵建立的模型在T4溶酶菌突变体的预测效果准确率上达到85%。我们的置换矩阵包含了丰富的突变分析信息。最后，他们提出一个可视化方法来检查每一个原子对分类决定的贡献。

#### conclusions
端对端的深度学习模型比传统的手工设计特征方法表现得好，说明了3DCNN框架能够用在蛋白质微环境分析和对其他蛋白质结构分析也有用。

### background
蛋白质位点是蛋白质结构里的微环境，区分于他们的结构和功能角色。一个位点能够用一个三维位置和这个位置的结构和功能所在的本地领域所定义。合理的蛋白质工程的核心是理解氨基酸的结构安排能够在蛋白质位点上创造出什么功能特性。

蛋白质里氨基酸的结构和功能角色的测定能够提供信息并帮助设计和修改蛋白质功能。找到在功能或者结构上重要的氨基酸能够将工程注意力集中，比如定向位点突变形成来改变特定蛋白质的功能特性。同时，这些知识能够帮助我们避免设计出会废除掉想要的功能的想法。传统地来说，实验的突变分析是用来确定改变单个氨基酸的效果。比如，在丙氨酸扫描里，蛋白质里的每一个氨基酸突变成丙氨酸，并且记录下相对应的功能和结构作用来识别出重要的氨基酸。这个技术经常用在蛋白质-蛋白质交互热点检测，用来识别交互的残基。然而这些实验方法是费时间的劳动力的。另外，也没有在氨基酸在哪些位置会耐受的信息。

蛋白质结构数据的增加给了我们机会可以系统性地学习潜在模式。任何一个计算性的蛋白质分析就是如何展现蛋白质结构信息。机器学习方法的效果更多的取决于数据的表达方式，而不是机器学习算法。好的特征表达能够有效率的抓到重要的信息，但是不好的会创造有干扰信息切没有潜在模式的分布。大部分方法依靠于根据蛋白质稳定性和化学成分构成手动提取的特征。比如，以特性为基础的表征...

蛋白质工程就是通过对蛋白质化学，蛋白质晶体学和蛋白质动力学的研究，获得有关蛋白质理化特性和分子特性的信息，在此基础上对编码蛋白质的基因进行有目的的设计和改造，通过基因工程技术获得可以表达蛋白质的转基因生物系统。

以蛋白质分子的结构规律及其生物功能的关系作为基础，通过化学，物理和分子生物学的手段进行基因修饰或基因合成，对现有蛋白质进行改造，或制造一种新的蛋白质，以满足人类对生产和生活的需求。

**研究内容**
1. 蛋白质结构分析
    蛋白质工程的核心内容之一就是收集大量的蛋白质分子结构的信息，以便建立结构与功能之间关系的数据库，为蛋白质结构与功能之间关系的理论研究奠定基础。三维空间结构的测定是验证蛋白质设计的假设即证明是新结构改变了原有生物功能的必需手段。
2. 结构，功能的设计和预测
    根据对天然蛋白质结构与功能分析建立起来的数据库的数据，可以预测一定氨基酸序列肽链空间结构和生物功能；
3. 创造或改造蛋白质--新蛋白质
    蛋白质的改造，从简单的物理，化学法到复杂的基因重组等等有多重方法。
    1. 物理，化学法：对蛋白质进行变形，复性处理，修饰蛋白质侧链官能团，分割肽链，改变表面电荷分布促进蛋白质形成一定的立体构象等等；
    2. 生物化学法：使用蛋白酶选择性地分割蛋白质，利用转糖苷酶，酯酶，酰酶等去除或连接不同化学基因，利用转酰胺酶使蛋白质发生胶连等等。以上方法只能对同或相似的基团或化学键发生作用，缺乏特异性，不能针对特定的部位起作用。
    3. 采用基因重组技术或人工合成DNA，不但可以改造蛋白质而且可以实现从头合成全新的蛋白质。

蛋白质是由不同的氨基酸按一定的顺序通过肽键连接而成的肽构成的。氨基酸序列就是蛋白质的一级结构，它决定着蛋白质的空间结构和生物功能。而氨基酸序列是由合成蛋白质的基因的DNA序列决定的，改变DNA序列就可以改变蛋白质的氨基酸序列，实现蛋白质的可调控生物合成。

定位突变蛋白质中的氨基酸是由基因中的三联密码决定的，只要改变其中的一个或两个就可以改变氨基酸。通常改变某个位置的氨基酸，研究蛋白质结构，稳定性或催化特性。

# 结论

从2017年开始有研究讨论用机器学习知识帮助蛋白质突变分析。例如用3DCNN预测对于某一个位点最合适的氨基酸或者指导蛋白质突变获得功能。

2017年，Torng and Altman提出使用3DCNN框架来做蛋白质分析。使用深度神经网络的好处是它能够自动的提取出对任务有用的特征，不需要我们对特定的任务作出假设和前验知识。他们用3DCNN模型来区分20个氨基酸的微环境，从而达到能够预测哪一个氨基酸对某一个位置是最合适的目的。另外他们的预测统计分析提供了每一个氨基酸被另一个氨基酸替代的概率信息。根据这些信息得到的替换矩阵，也通过实验证明他们包含跟多的信息。另外他们也提供了一种可视化技术，能够将每一个原子对最终决策的贡献可视化。他们的成功说明了用深度神经网络来做蛋白质结构分析是可行的。

2020年，Shroff et al.也提出了用3DCNN框架来识别出功能获得性的突变。他们在Torng and Altman的模型的基础上，他们做了一些改变来增加深度神经网络和氨基酸周围化学成分的联系。他们加入了氢原子，局部电荷，溶解可及性 的相关信息。另外他们修改了数据集以此免除掉了一些误差。并且，增加了半胱氨酸和甲硫氨酸的频率来提高准确率。

2022年，Hongyuan Lu et al.，也用了3DCNN网络来预测突变。他们通过此方法得到的FAST-PETase能在（30-50）度的反应条件下，针对51种低结晶度的PET产品包装，预处理过的高结晶度PET塑料瓶，以及化纤织物等，均能直接产生快速高效的解聚效果。

## 方向分类
rational and random mutate enzyme, entail evolution, selection and screening.




蛋白质工程是通过物理，化学，生物和基因重组等技术改造蛋白质或设计合成具有特定功能的新蛋白质。利用基因工程的手段，在目标蛋白的氨基酸列上引入突变，从而改变目标蛋白的空间结构，最终达到改善起功能的目的。
传统的蛋白质突变形成可以大致分成两类，基于理性的和随机的（也可以叫他们定向位点突变和基于演变的方法）。

随机的方法的缺点是无法预测为什么所选择的氨基酸替换会导致想要的改变。这些方法只需要对结构和功能有一点了解就行。这些方法的成功依靠于筛选甄别方法的效率和可靠。最简单的随机方法是饱和诱变，就是将蛋白质的某一个氨基酸替换成其他的任意氨基酸。[1,3]

理性的方法通过一些机制和知识挑选氨基酸，在特定的位点上。这些方法能够减少需要测试的变体，节省时间和人力。[2]

定向位点需要对结构和功能有深的了解，通过物理，生物，化学等知识来做选择。然而传统的做法是费时间和费力的。随着蛋白质结构数据的增加，我们能够能够使用机器学习的方法来从数据中获取潜在的信息。[4,5,6,7,8,9,10]

随着深度学习在其他领域的成功，3DCNN也被用来从蛋白质3D结构中提取信息来预测位点突变，并且取得了成功。[11,12,13]

利用机器学习方法很大取决于对特征的选择。基本的传统方法都适用物理化学的特性来做特征 [15,16,17,18] 。但是手动设计特征很耗时间和人力。深度神经网络的好处就是我们不需要对特征做任何的处理和对特征不需要有前验知识和假设。
