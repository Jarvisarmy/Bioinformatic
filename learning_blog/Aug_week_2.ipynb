2. **Schreiber et al., 2020, Ledidi: Designing genomic edits that induce functional activity**
    * Ledidi, an approach that treats the design of genomic edits as an optimization problem where the goal is to produce the desired output from a predictive model. 
    * use Gumbel-Softmax reparameterization trick to solve the discrete nature
    * the objective function contains two parts
        1. the distance between the original sequence and the proposed edited sequence (the sequence loss)
        2. the distance between the model output using the edited sequence and the desired output (the functional loss)
        3. the second term is weighted by $\lambda$ to either encourage a small number of edits when small or a closer match to the desired output when large
    * The main idea is about how to reverse the predictive model to find a minimal set of mutations to achieve desirable targetsm

# 总结
随着深度神经网络在其他方面的成功，很多人也把深度模型应用到了生物信息中，比如蛋白质工程和基因工程。

深度神经网络在蛋白质工程中的用法可以大致分成两类，一类是直接预测突变位点和需要替换的氨基酸。这一类通常分为两步，第一步是知道哪个点不合适，第二步是解决应该用什么氨基酸替换。另一类是给定蛋白质的氨基酸序列或者3D结构，预测突变的效果，比如预测蛋白质-蛋白质相互作用，药物-靶点相互作用，稳定性，热稳定性，能量变化，指定功能等等。第二类方法通常需要跟其他方法结合，比如screening，evolution等得到突变的蛋白质，再用特定模型来预测其突变特性。相比于第一类，这些方法还是比较废时间和人力的。

从输入信息来分类的话，sequence-based和structured-based. sequence-based 是利用蛋白质的氨基酸序列来做分析。structured-based是通过蛋白质的蛋白质结构（碳，氢，氧，氮等重要的生物大分子）,通常指3级结构，大部分来源于PDB。蛋白质的三级结构包含了更多的信息且其功能通常由3D结构决定。但这也意味着我们需要处理更多的信息和难度。另外一方面从数据数量上来看，跟氨基酸序列相比，我们所知道的蛋白质结构是相对较少的，随着AlphaFold2的成功，他能预测出基本上所有已知的蛋白质结构，极大的推进了structured-based的研究。但是预测的是有错误率的，在这个基础上进行预测，对准确率还是有影响。另外20个常见氨基酸的排列组合可以说有无数种，但我们一直的常见的或者特定目标的氨基酸序列是少量且固定的，意味着通过氨基酸序列来预测他们的突变位点或者突变效果是有效且可行的。也有一些人同时运用这两种信息来做预测。

从深度模型上，目前大部分模型都适用CNN, 包括3DCNN, 2DCNN, residual CNN.
