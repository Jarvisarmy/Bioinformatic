```python
import pandas as pd
import numpy as np
```

```python
def convert_stabilizing(x):
    try: 
        if str(x) == "False":
            return "False"
        else:
            return "True"
    except:
        return ''
```

```python
data = pd.read_csv('data/K50_dG_Dataset1_Dataset2.csv',dtype={"WT_name":str,"Stabilizing_mut":str})
```

```python
data
```

```python
data.columns
```

```python
attributes = ["name","WT_name","aa_seq","ddG_ML"] #"mut_type"
qualified_data = data.loc[(data['Stabilizing_mut'] == "True") & (data['ddG_ML']!='-') &(data['WT_name']!=data['name'])]
```

```python
# 获得所有wild type 的sequences
WT_sequences = data.loc[data['name'] == data['WT_name']][['name','aa_seq']].groupby('name').first()
WT_sequences
```

```python
# 总共有228个 wild type sequences, 可以放进去纠错模型训练
qualified_data[attributes].groupby('WT_name').first()
```

```python
# 每一本wild type sequence有多个可能的突变体，总共有2263组数据
selected_data = qualified_data[attributes]
```

```python
selected_data.insert(selected_data.shape[1],'WT_seq',"")
```

```python
for idx, row in selected_data.iterrows():
    row['WT_seq'] = WT_sequences.loc[row['WT_name']].aa_seq
```
